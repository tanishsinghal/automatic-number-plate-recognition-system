```python
import cv2 # make sure to import the cv2 module

img = cv2.imread('/test.jpg')

# Check if the image was loaded correctly
if img is None:
    print("Error: Could not load image. Please check the file path and ensure the image exists.")
else:
    gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    print(gray)
```

# New Section

```python

gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
```

```python

plt.imshow(gray)
```

```python

plt.imshow(cv2.cvtColor(gray, cv2.COLOR_BGR2RGB))
```

```python

bfilter = cv2.bilateralFilter(gray, 11, 11, 17)
```

```python

edged = cv2.Canny(bfilter, 30, 200)
```

```python
plt.imshow(cv2.cvtColor(edged, cv2.COLOR_BGR2RGB))
```

```python
#  cv.CHAIN_APPROX_NONE gives (734 points) and  with cv.CHAIN_APPROX_SIMPLE (only 4 points)

keypoints = cv2.findContours(edged.copy(), cv2.RETR_TREE, cv2.CHAIN_APPROX_SIMPLE)
```

```python

# Simplifies how contonours are actually returned
contours = imutils.grab_contours(keypoints)
```

```python

contours = sorted(contours, key = cv2.contourArea, reverse = True)[:10]
```

```python

print(contours)
```

```python
location = None

for contour in contours:
  # cv2.approxPolyDP returns a resampled contour, so this will still return a set of (x, y) points
  approx = cv2.approxPolyDP(contour, 10, True)
  if len(approx) == 4:
    location = approx
    break
```

```python

print(location)
```

```python

mask = np.zeros(gray.shape, np.uint8)
```

```python

new_image = cv2.drawContours(mask, [location], 0, 255, -1)

new_image = cv2.bitwise_and(img, img, mask = mask)


plt.imshow(cv2.cvtColor(new_image, cv2.COLOR_BGR2RGB))
```

```python
(x, y) = np.where(mask == 255)

(x1, y1) = (np.min(x), np.min(y))

(x2, y2) = (np.max(x), np.max(y))
```

```python
cropped_image = gray[x1:x2+3, y1:y2+3]

plt.imshow(cv2.cvtColor(cropped_image, cv2.COLOR_BGR2RGB))
```
